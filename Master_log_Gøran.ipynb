# Lammps commands

How is a LAMMPS input script formatted and how to use the input script commands to define a LAMMPS simulation

### LAMMPS input scripts

```python
from IPython.display import Latex
```

(1) The ordering of commands is in most cases not important. The input script is read one line at a time, and each command takes effect when read. Thus the sequence of commands:

\begin{verbatim}
timestep  0.5
run       100
run       100
\end{verbatim}

timestep is in $ fmsec $, and two simulations is run, each with a 100 timesteps. If you switch places on line one and two, a simulation with 100 timesteps will run with the default timestep $ 1.0 fmsec $, and then the timestep will be set to $ 0.5 fmsec $ before the next simulation is executed.

(2) Some commands are only valid if they follow other commands. For example you cannot set the temperature of a group of atoms until atoms have been defined, and a group command is used to define which atoms belong to the group. The read_data command initializes the system by setting up the simulation box and assign atoms to processors. If default values are not desired, the processor and boundary commands need to be used before read_data to tell LAMMPS how to map processors to the simulation box.

Many input script errors are detected by LAMMPS, and an ERROR or WARNING messaage is printed.

# LAMMPS in.portlandite

What kind of potential should we use between the different particles? The standard Lenny-Jones potential is formulated as $$ E = 4\epsilon \Big[ \big( \frac{\sigma }{r} \big) ^{12} - \big( \frac{\sigma }{r} \big) ^6 \Big] \ \  for \  \ r < r_c $$, where $ r_c $ is the cut-off lenght for the potential. It is normal to give a cut-off length at which point the LJ-potential is in practise equal to zero. 

The abow equation for the standard Lenny-Jones potential is in Cygan et al. (Molecular Models of Hydroxides, Oxyhydroxides and Clay Phases and the development of a general Force Field) given as: 
\begin{equation}
E_{VDW} = \sum _{i\neq j} D_{O,ij}\Big[ \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{12} - \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{6} \Big]
\end{equation}

where the empirical parameters for the cut-off distance parameter $ R_{O,ij} $ and the energy parameter $ D_{O,ij} $ is given as the arithmetic mean and geometric mean respectively. Hereunder I present the obtained values for the nonbond values between $ Ca $ and $ O $:
\begin{equation}
R_{O,ij} = \frac{1}{2}(R_{O,i} + R_{O,j}) = 0.5(6.2428+3.5532) = 4.898
\end{equation}

\begin{equation}
D_{O,ij} = \sqrt{D_{O,i}D_{O,j}} = \sqrt{0.1554\cdot 5.0298e-6} = 8.841e-4
\end{equation}

(values are obtained from table 1 in Cygan et al. for the Ca-O coefficients. These are the nonbond parameters used in the pair_style and pair_coeff commands in LAMMPS) We obtain coefficients as:
\begin{equation}
R_{O,Caca} = 6.428
\end{equation}
\begin{equation}
D_{O,CaCa}  = 5.0298E-6
\end{equation}

for calcium and calcium, and for oxygen and oxygen we get:
\begin{equation}
R_{O,OxOx} = 3.5532
\end{equation}
\begin{equation}
D_{O,OxOx}  = 0.1554
\end{equation}

We write the results into LAMMPS as:

#pair_coeff          i j $ D_{O,ij} $ $ R_{O,ij} $ cutoff_lj cutoff_coul

pair_coeff          1 1 5.0298E-6 6.428 3.0 3.0

pair_coeff          1 2 8.841e-4 4.898 3.0 3.0

and so on...


## Potentials

### The Lennard-Jones potential. In LAMMPS named E_vdw for van der Wals

\begin{equation}
E_{VDW} = \sum _{i\neq j} D_{O,ij}\Big[ \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{12} - \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{6} \Big]
\end{equation}

### The Coulomb interaction

\begin{equation}
E_{coul} = \frac{e^2}{4\pi \epsilon _0}\sum _{i\neq j} \frac{q_i q_j}{r_{ij}}
\end{equation}

### Bond interaction

\begin{equation}
E_{bond} = k_1(r_{ij}-r_o)^2
\end{equation}

### Angle interaction

\begin{equation}
E_{angle} = k_2(\theta _{ijk} - \theta _o)^2
\end{equation}

```python
filename = open("/home/goran/lammps-28Jun14/examples/portlandite/run_2014_09_16/dump.portlandite_restart","r")
line1 = filename.readline()
line2 = filename.readline()
Nbonds = 1600
Natoms = 4000
N_Ca = (4.0/20)*Natoms # = 800
N_O = (8.0/20)*Natoms  # =1600
N_H = N_O
Time = []
Temp1 = []
Temp2 = []
dCa = []
dO = []
dH = []
Bondl = []
for line in filename:
    TimeStep, myTemp, TempCOM, dispCa, dispO, dispH, bondlOH = line.split()
    Time.append(float(TimeStep)), Temp1.append(float(myTemp)), Temp2.append(float(TempCOM))
    dCa.append(float(dispCa)/N_Ca), dO.append(float(dispO)/N_O),dH.append(float(dispH)/N_H),Bondl.append(float(bondlOH)/Nbonds)

    
length = len(dCa)
start = int(length/10)

import numpy as np
std_Ca = np.std(dCa[start:])
std_O = np.std(dO[start:])
std_H = np.std(dH[start:])
std_bondl = np.std(Bondl[start:])
std_temp = np.std(Temp2[start:])
mean_Ca = np.mean(dCa[start:])
mean_O = np.mean(dO[start:])
mean_H = np.mean(dH[start:])
mean_bondl = np.mean(Bondl[start:])
mean_temp = np.mean(Temp2[start:])

xmin = Time[0]
xmax = Time[-1]
ymin = 290
ymax = 310
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
#plt.subplot(1,3,1)
plt.plot(Time,Temp2,'r-')
plt.legend(['temp/com'], loc='upper right')
plt.title('Temperature as a function of time \n T = %.1f +- %.1f K' % (mean_temp, std_temp))
plt.xlabel('Time [fsec]')
plt.ylabel('Temperature [K]')
plt.axis([xmin,xmax,ymin,ymax])


ymax = 1.024
ymin = 1.022
f2 = plt.figure()
#plt.subplot(1,3,2)
plt.plot(Time, Bondl,'r-')
plt.legend(['OH(t)'])
plt.title('OH bondlength as a function of time \n Bondlength OH = %.4f +- %.4f' % (mean_bondl, std_bondl))
plt.xlabel('Time [fsec]')
plt.ylabel('Bondlength [Aangstrom]')
plt.axis([xmin,xmax,ymin,ymax])

ymin = 1.5
ymax = 4.0
f3 = plt.figure()
plt.plot(Time, dCa, 'b-')
plt.hold(True)
plt.plot(Time, dO, 'r-')
plt.plot(Time, dH, 'g-')
plt.hold(False)
plt.title('Displacement of particles as function of time \n Ca = %.2f +- %.3f \n O = %.2f +- %.3f \n H = %.2f +- %.3f ' % (mean_Ca, std_Ca, mean_O, std_O, mean_H, std_H))
plt.legend(['Ca','O','H'],loc='upper right')
plt.xlabel('Time [fsec]')
plt.ylabel('Displacement [Aangstrom]')
plt.axis([xmin,xmax,ymin,ymax])

plt.show()

```

## Modeling water

I have downloaded files from http://www.orsi.sems.qmul.ac.uk/downloads.html . Although much of the code was wirking fine, the fix shake is an unphysical implementation that neglects the wibrations in the water molecule, I removed this, and had to use a smaller timestep $ \Delta t = 0.1 $. Running a npt simulation of 10x10x10 water molecules for 200000 timesteps, which corresponds to $20000fsec$, I gained some results that can be used to verify the model. The simulation were done at fixing the temperature at $273K$ and keeping the external pressure at $1atm$. We see that the bondlenght of OH is consistent with literature, just some above $1.0Angstrom$, and that the density of water at about $298K$ is about $0.56g/cm^3 = 560\frac{kg}{m^3} $. But from litterature we know that the density of water at zero degrees celcius chould be $1000\frac{kg}{m^3}$, at 25degrees celcius we should find a density of about $997\frac{kg}{m^3}$. Why do we only get half the density?

```python
filename = open("/home/goran/lammps-28Jun14/examples/water/dump.water_LAMMPS","r")
line1 = filename.readline()
line2 = filename.readline()

timestep = 0.1 # fsec
Time = []
Temp1 = []
Temp2 = []
msd = []
rog = []
dens = []
Bondl = []
for line in filename:
    TimeStep, myTemp, TempCOM, MSD, ROG, bondlOH, Dens = line.split()
    Time.append(float(TimeStep)*timestep), Temp1.append(float(myTemp)), Temp2.append(float(TempCOM))
    msd.append(float(MSD)), rog.append(float(ROG)), Bondl.append(float(bondlOH)), dens.append(float(Dens))
 
length = len(Time)
start = (int(length/10))

import numpy as np
std_msd = np.std(msd[start:])
std_rog = np.std(rog[start:])
std_dens = np.std(dens[start:])
std_bondl = np.std(Bondl[start:])
std_temp = np.std(Temp2[start:])
mean_msd = np.mean(msd[start:])
mean_rog = np.mean(rog[start:])
mean_dens = np.mean(dens[start:])
mean_bondl = np.mean(Bondl[start:])
mean_temp = np.mean(Temp2[start:])

xmin = Time[0]
xmax = Time[-1]
ymin = 100 # Kelvin
ymax = 1000
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
#plt.subplot(1,3,1)
plt.plot(Time,Temp2,'r-')
plt.legend(['temp/com'], loc='upper right')
plt.title('Temperature as a function of time \n T = %.1f +- %.1f K' % (mean_temp, std_temp))
plt.xlabel('Time [fsec]')
plt.ylabel('Temperature [K]')
plt.axis([xmin,xmax,ymin,ymax])


ymax = 1.024
ymin = 1.0
f2 = plt.figure()
#plt.subplot(1,3,2)
plt.plot(Time, Bondl,'r-')
plt.legend(['OH(t)'])
plt.title('OH bondlength as a function of time \n Bondlength OH = %.4f +- %.4f' % (mean_bondl, std_bondl))
plt.xlabel('Time [fsec]')
plt.ylabel('Bondlength [Aangstrom]')
plt.axis([xmin,xmax,ymin,ymax])

ymin = 0
ymax = 140
f3 = plt.figure()
plt.plot(Time, msd, 'b-')
plt.title('Mean square displacement of water molecules')
#plt.title('Mean square displacement of water molecules \n msd = %.2f +- %.3f' % (mean_msd, std_msd))
plt.legend(['msd(t)'], loc='lower right')
plt.xlabel('Time [fsec]')
plt.ylabel('Displacement [Aangstrom]')
plt.axis([xmin,xmax,ymin,ymax])

ymin = 15
ymax = 25
f4 = plt.figure()
plt.plot(Time, rog, 'b-')
plt.title('radius of gyration for the water molecules')
#plt.title('radius of gyration for the water molecules \n rog = %.2f +- %.3f' % (mean_rog, std_rog))
plt.legend(['rog(t)'], loc='lower right')
plt.xlabel('Time [fsec]')
plt.ylabel('rog [Aangstrom]')
plt.axis([xmin,xmax,ymin,ymax])

ymin = 0.55
ymax = 0.7
f5 = plt.figure()
plt.plot(Time, dens, 'b-')
plt.title('Density of water as function of time \n rho = %.2f +- %.3f ' % (mean_dens, std_dens))
plt.legend(['rho'],loc='upper right')
plt.xlabel('Time [fsec]')
plt.ylabel('rho [g/cm^3]')
plt.axis([xmin,xmax,ymin,ymax])

plt.show()
```
