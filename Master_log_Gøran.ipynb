# Lammps commands

How is a LAMMPS input script formatted and how to use the input script commands to define a LAMMPS simulation

### LAMMPS input scripts

```python
from IPython.display import Latex
```

(1) The ordering of commands is in most cases not important. The input script is read one line at a time, and each command takes effect when read. Thus the sequence of commands:

\begin{verbatim}
timestep  0.5
run       100
run       100
\end{verbatim}

timestep is in $ fmsec $, and two simulations is run, each with a 100 timesteps. If you switch places on line one and two, a simulation with 100 timesteps will run with the default timestep $ 1.0 fmsec $, and then the timestep will be set to $ 0.5 fmsec $ before the next simulation is executed.

(2) Some commands are only valid if they follow other commands. For example you cannot set the temperature of a group of atoms until atoms have been defined, and a group command is used to define which atoms belong to the group. The read_data command initializes the system by setting up the simulation box and assign atoms to processors. If default values are not desired, the processor and boundary commands need to be used before read_data to tell LAMMPS how to map processors to the simulation box.

Many input script errors are detected by LAMMPS, and an ERROR or WARNING messaage is printed.

# LAMMPS in.portlandite

What kind of potential should we use between the different particles? The standard Lenny-Jones potential is formulated as $$ E = 4\epsilon \Big[ \big( \frac{\sigma }{r} \big) ^{12} - \big( \frac{\sigma }{r} \big) ^6 \Big] \ \  for \  \ r < r_c $$, where $ r_c $ is the cut-off lenght for the potential. It is normal to give a cut-off length at which point the LJ-potential is in practise equal to zero. 

The abow equation for the standard Lenny-Jones potential is in Cygan et al. (Molecular Models of Hydroxides, Oxyhydroxides and Clay Phases and the development of a general Force Field) given as: 
\begin{equation}
E_{VDW} = \sum _{i\neq j} D_{O,ij}\Big[ \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{12} - \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{6} \Big]
\end{equation}

where the empirical parameters for the cut-off distance parameter $ R_{O,ij} $ and the energy parameter $ D_{O,ij} $ is given as the arithmetic mean and geometric mean respectively. Hereunder I present the obtained values for the nonbond values between $ Ca $ and $ O $:
\begin{equation}
R_{O,ij} = \frac{1}{2}(R_{O,i} + R_{O,j}) = 0.5(6.2428+3.5532) = 4.898
\end{equation}

\begin{equation}
D_{O,ij} = \sqrt{D_{O,i}D_{O,j}} = \sqrt{0.1554\cdot 5.0298e-6} = 8.841e-4
\end{equation}

(values are obtained from table 1 in Cygan et al. for the Ca-O coefficients. These are the nonbond parameters used in the pair_style and pair_coeff commands in LAMMPS) We obtain coefficients as:
\begin{equation}
R_{O,Caca} = 6.428
\end{equation}
\begin{equation}
D_{O,CaCa}  = 5.0298E-6
\end{equation}

for calcium and calcium, and for oxygen and oxygen we get:
\begin{equation}
R_{O,OxOx} = 3.5532
\end{equation}
\begin{equation}
D_{O,OxOx}  = 0.1554
\end{equation}

We write the results into LAMMPS as:

#pair_coeff          i j $ D_{O,ij} $ $ R_{O,ij} $ cutoff_lj cutoff_coul

pair_coeff          1 1 5.0298E-6 6.428 3.0 3.0

pair_coeff          1 2 8.841e-4 4.898 3.0 3.0

and so on...


## Potentials

### The Lennard-Jones potential. In LAMMPS named E_vdw for van der Wals

\begin{equation}
E_{VDW} = \sum _{i\neq j} D_{O,ij}\Big[ \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{12} - \Big( \frac{R_{O,ij}}{r_{ij}} \Big) ^{6} \Big]
\end{equation}

### The Coulomb interaction

\begin{equation}
E_{coul} = \frac{e^2}{4\pi \epsilon _0}\sum _{i\neq j} \frac{q_i q_j}{r_{ij}}
\end{equation}

### Bond interaction

\begin{equation}
E_{bond} = k_1(r_{ij}-r_o)^2
\end{equation}

### Angle interaction

\begin{equation}
E_{angle} = k_2(\theta _{ijk} - \theta _o)^2
\end{equation}
